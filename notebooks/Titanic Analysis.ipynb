```python
# Multiple aggregations: survival rate and mean age by class and gender
df.groupby(['Pclass', 'Sex']).agg({'Survived': 'mean', 'Age': 'mean'})
# For 1st class - the survival rate of the females is 96% with the average age of 34 approximately.
# The survival rate of the males is 38% with the average age of 39
# For 2nd class - the survival rate of the females is 92% with the average age of 29
# While that of the males is 16% with the average age of 31
# For 3rd class - the survival rate of the females is 50% with the average age of 24
# While that of the males is 14% with the average age of 27
```

```python
# What is the average fare of females in 1st class compared to males in 1st class?

first_class = df.groupby(['Pclass','Sex'])['Fare'].mean().loc[1]
first_class
```

```python
import matplotlib.pyplot as plt
first_class.plot(
    kind = "bar",
    color = ["purple", "orange"],
    title = "Average Fare in First Class by Gender"
)
plt.ylabel("Average Fare")
plt.xlabel("Sex")
plt.xticks(rotation = 0)
plt.grid(axis = "y", linestyle = "--", alpha = 0.7)
plt.show()
```

```python
df["age_group"] = df["Age"].apply(lambda x: "Child" if x < 18 else "Adult")

# Calculate survival rate by age group
survival_by_age_group = (
    df
    .groupby("age_group")["Survived"]
    .mean()
    .reset_index()
)

survival_by_age_group
```

```python
import matplotlib.pyplot as plt

plt.bar(
    survival_by_age_group["age_group"],
    survival_by_age_group["Survived"]
)

plt.title("Survival Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.show()
```

*Insight*


Children exhibited higher survival rates compared to adults, suggesting prioritization during evacuation efforts.

```python
df['Family_size'].value_counts()
```

```python
def family_group(size):
    if size == 0:
        return "Solo"
    elif size <= 3:
        return "Small Family (1–3)"
    else:
        return "Large Family (4+)"

df["family_group"] = df["Family_size"].apply(family_group)

survival_by_family_group = (
    df.groupby("family_group")["Survived"]
    .mean()
    .reset_index()
)

survival_by_family_group
```

```python
plt.bar(
    survival_by_family_group["family_group"],
    survival_by_family_group["Survived"]
)

plt.title("Survival Rate by Family Group")
plt.xlabel("Family Group")
plt.ylabel("Survival Rate")
plt.show()
```

*Insight*

Passengers traveling with small to medium family sizes exhibited higher survival rates compared to those traveling with very large families

```python
df.head(2)
```

```python
# Create fare bins (quartiles)
df["fare_group"] = pd.qcut(
    df["Fare"],
    q=4,
    labels=["Low Fare", "Mid-Low Fare", "Mid-High Fare", "High Fare"]
)

# Calculate survival rate by fare group
survival_by_fare = (
    df.groupby("fare_group")["Survived"]
    .mean()
    .reset_index()
)

survival_by_fare
```

```python
plt.bar(
    survival_by_fare["fare_group"],
    survival_by_fare["Survived"]
)

plt.title("Survival Rate by Fare Level")
plt.xlabel("Fare Group")
plt.ylabel("Survival Rate")
plt.show()
```

*Insight*


Passengers who paid higher fares generally exhibited higher survival rates, indicating a relationship between socio-economic status and survival outcomes.

Dealing with Outliers

```python
# Outliers are observations in a data distribution that don't fit in the overall distribution or spread of data
# Are extreme data that do not follow normal pattern.
# They exist because of inherent varianlitybthat exist in data distribution
# Maybe an entry error, measurement error or natural occurence
```

```python
data_Titanic = pd.read_csv('Titanic-Dataset.csv')
data_Titanic.head(2)
```

```python
# Featured column
data_Titanic['Age'].head()
```

```python
# import visualization libraries
import seaborn as sns
```

```python
# Box plot
sns.boxplot(data = data_Titanic, x = 'Age')
```

```python
data_Titanic['Age'].describe()
```

```python
# Distribution plot
sns.displot(data = data_Titanic, x = 'Age', bins = 10, kde = False) #KDE - kernel density estimate
```

```python
# Scatter plot - relationship
sns.scatterplot(data = data_Titanic, x = 'Age', y = 'Fare', hue = 'Sex')
```

```python
# Tukey's rule 
# Q1 - 1.5 * IQR - Lower limit
# Q3 + 1.5 * IQR - Upper limit
```

```python
Q1 = data_Titanic['Age'].quantile(.25)
Q3 = data_Titanic['Age'].quantile(.75)
IQR = Q3 - Q1
```

```python
IQR
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python
lower_limit
```

```python
upper_limit
```

```python
outlier_upper = (data_Titanic['Age'] > upper_limit)
```

```python
outlier_upper
```

```python
len(data_Titanic['Age']) - len (data_Titanic['Age'][outlier_upper]) # The eleven values are our supposed outliers
```

```python
data_Titanic_N = data_Titanic[-(outlier_upper)]
```

```python
data_Titanic_N.shape
```

```python
data_Titanic['Age'][outlier_upper]
```

```python
# Visualize data without outliers
sns.boxplot(data = data_Titanic_N, x = 'Age')
```

```python
sns.displot(data = data_Titanic_N, x = 'Age', bins = 10, kde = False)
```

```python
sns.scatterplot(data = data_Titanic_N, x = 'Age', y = 'Fare', hue = 'Sex')
```

## Key Insights

- Female passengers had significantly higher survival rates compared to male passengers.
- Passengers in first class were more likely to survive than those in lower classes.
- Children had better survival outcomes than adults.
- Passengers traveling with small to medium family sizes exhibited higher survival rates compared to those traveling with very large families.
- Higher fare payments were generally associated with improved survival chances.


## Implications

- Emergency evacuation protocols should prioritize vulnerable groups.
- Safety measures should ensure equitable access across socio-economic classes.
- Passenger group dynamics should be considered in emergency response planning.


## Limitations

- The analysis is based on historical data and does not imply causation.
- Some features contain missing values, which may affect results.
- Fare and passenger class are strongly correlated, which may influence interpretations.
